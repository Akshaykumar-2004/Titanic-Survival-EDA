```python
#cleaning process
df.isnull().sum()
```

```python
#fixing the missing values
df.drop('Cabin',axis=1,inplace=True) #because cabin have too many missing values and also not useful for our analysis

```

```python
median_age = df['Age'].median() #calculating medain of age .since age is a relvant and useful data for our analysis
df['Age'].fillna(median_age,inplace=True) #filling the missing values
```

```python
most_common_port = df['Embarked'].mode()[0]#checks most repetad value
df['Embarked'].fillna(most_common_port,inplace=True) #filling values in embarked col
```

```python
df.isnull().sum() #checking the dataset for missing valyues are filled or not
```

```python
#feature engineering .creating new clues with existing clues
df['FamilySize']=df['SibSp'] + df['Parch'] + 1 #created new col
```

```python
#droping col which are not needed now
df.drop(['PassengerId','Name','Ticket','SibSp','Parch'], axis=1, inplace=True)
```

```python
df.head() #after this much analysis looking our data
```

```python
#vizualization how many people survived univarite analysys
survival_counts =df['Survived'].value_counts()
print(survival_counts) #total people survived
plt.figure(figsize=(6,6))
plt.bar(survival_counts.index,survival_counts.values)
plt.xlabel('Survived')
plt.ylabel('number of passengers')
plt.title('Survival Count')
plt.show()   #from this we understand that 549 people died
```

```python
#does ticket class matter for survival (bivariete analysis)
class_survival=pd.crosstab(df['Pclass'],df['Survived'])
print(class_survival)

```

```python
#vizualization
class_survival.plot(kind='bar')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0, ticks=[0,1,2], labels=['1st Class', '2nd Class', '3rd Class'])
plt.legend(['Did Not Survive', 'Survived'])
plt.show()
```

```python
#does gender matter in survival
gender_survival=pd.crosstab(df['Sex'],df['Survived'])
print(gender_survival)
```

```python
#visulization

gender_survival.plot(kind='bar')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.legend(['Did Not Survive', 'Survived'])
plt.show()
```

```python
#what about age is this matter
# Create a figure with two subplots side-by-side
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot histogram for those who did not survive
ax[0].hist(df[df['Survived']==0]['Age'], bins=20, color='red')
ax[0].set_title('Age Distribution of Non-Survivors')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Number of Passengers')

# Plot histogram for those who survived
ax[1].hist(df[df['Survived']==1]['Age'], bins=20, color='green')
ax[1].set_title('Age Distribution of Survivors')
ax[1].set_xlabel('Age')

plt.show()
```
